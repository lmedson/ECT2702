```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
import seaborn as sb
#%matplotlib.inline
# old version
%pylab inline 
```

```python
df = pd.read_csv('data/iris.csv')
df.head()
```

```python
sb.pairplot(df, hue = "target")
```

```python
sb.pairplot(df)
```

```python
#features do objeto
X = np.array(df.drop('target',axis=1))
```

```python
X
```

## K-Means

```python
# from sklearn.cluster import Kmeans
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
```

```python
kmeans.fit(X)
```

```python
kmeans.labels_
```

# Comparando classes criadas com classes existentes de iris

```python
df['K-classes'] = kmeans.labels_
```

```python
df
```

```python
sb.pairplot(df,hue='target')
```

```python
sb.pairplot(df,'K-classes')
```

```python
questions_ds = pd.read_csv('data/perfil_politico.csv')
```

```python
questions_ds.values
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2,init = 'random', random_state=1)
```

```python
kmeans.fit(questions_ds)
#kmeans.fit_predict(questions_ds.values) mesmo que usar, kmeans.labels_
```

```python
questions_ds['cluster'] = kmeans.labels_
```
